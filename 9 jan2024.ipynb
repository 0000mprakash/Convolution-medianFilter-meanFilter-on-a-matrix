```python
import random
import numpy as np
import matplotlib.pyplot as plt

a=np.zeros((12,12))
for i in range(12-2):
  for j in range(12-2):
    a[i+1][j+1]=random.randint(0,255)
b=np.empty((3,3))
for i in range(3):
  for j in range(3):
    b[i][j]=random.randint(0,1)

output=np.zeros((a.shape[0]-b.shape[0]+1,a.shape[1]-b.shape[1]+1))
for i in range(output.shape[0]):
  for j in range(output.shape[0]):
    sum=0
    for m in range(b.shape[0]):
      for n in range(b.shape[1]):
        sum+=b[m][n]*a[i+m][j+n]
    output[i][j]=sum

print(a)
print(b)
print(output)
  # submatrix = arr[np.ix_([0, 3], [1, 3])]
  # [i-1:i+1][j-1:j+1]
```

Mean Filter

```python
c=np.zeros((10,10))
for i in range(output.shape[0]):
  for j in range(output.shape[1]):
    sum=0
    for m in range(b.shape[0]):
      for n in range(b.shape[1]):
        sum+=b[m][n]*a[i+m][j+n]
    c[i][j]=sum/9

print(c)
```

Median Filter

```python
d=np.zeros((10,10))
array=[]
for i in range(output.shape[0]):
  for j in range(output.shape[1]):
    for m in range(b.shape[0]):
      for n in range(b.shape[1]):
        array.append(b[m][n]*a[i+m][j+n])
    array.sort()
    d[i][j]=array[len(array)//2]
    array.clear()

print(d)
```
